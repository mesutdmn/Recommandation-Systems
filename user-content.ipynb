```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def clean_data(data):
    data = data.copy()
    data = data.drop_duplicates()
    data = data.dropna()
    data = data.reset_index(drop=True)
    return data
```

```python
def user_recommend (userid, moviedf, ratingdf, corr=0.6):
    moviedf = moviedf.copy()
    ratingdf = ratingdf.copy()
    
    moviedf = clean_data(moviedf)
    ratingdf = clean_data(ratingdf)
    
    moviedf = moviedf.rename(columns={"id":"movieId"})
    moviedf.loc[:,"movieId"] = moviedf["movieId"].astype("int64")
    
    df = pd.merge(moviedf,ratingdf,on="movieId",how="inner")
    
    user_movie_df = df.groupby(["userId","title"])["rating"].mean().unstack()
    
    random_user_df = user_movie_df[user_movie_df.index==userid]
    
    movies_watched = random_user_df.dropna(axis=1).columns.tolist()
    lucky_guy_watched_ids = df["movieId"][(df["userId"]==userid)]
    movies_watched_df = user_movie_df[movies_watched]
    user_movie_count = movies_watched_df.notnull().sum(axis=1)
    
    filted_df = movies_watched_df[movies_watched_df.index.isin(user_movie_count.index)]
    corr_df = filted_df.T.corr().unstack()
    top_users = pd.DataFrame(corr_df[userid][corr_df[userid] > corr], columns=["corr"])
    
    top_users_ratings = pd.merge(top_users, ratings[["userId", "movieId", "rating"]], how='inner', on="userId")
    top_users_ratings = top_users_ratings[~(top_users_ratings["movieId"].isin(lucky_guy_watched_ids))]
    top_users_ratings['weighted_rating'] = top_users_ratings['corr'] * top_users_ratings['rating']
    recommendation_df = top_users_ratings.pivot_table(values="weighted_rating", index="movieId", aggfunc="mean")
    movies_to_be_recommend = recommendation_df[recommendation_df["weighted_rating"] > 3.5]
    recommend_by_user = pd.merge(movies_to_be_recommend,moviedf,how="inner",on="movieId")[["movieId","weighted_rating","title"]].sort_values(by="weighted_rating", ascending=False)
    return recommend_by_user
```

```python
def clear_doc(data,column):
    data = data.copy()
    data[column] = data[column].str.replace(r"[^\w\s]"," ",regex=True).str.replace(r"[\d]","",regex=True)
    return data[column]
```

```python
def content_recommend(movieId, moviedf, doc_column="overview"):
    moviedf = moviedf.copy()
    moviedf = clean_data(moviedf)
    
    moviedf = moviedf.rename(columns={"id":"movieId"})
    moviedf.loc[:,"movieId"] = moviedf["movieId"].astype("int64")
    
    moviedf[doc_column] = clear_doc(moviedf,doc_column)
    
    tfidf = TfidfVectorizer(stop_words="english", min_df=5)
    tfidf_matrix = tfidf.fit_transform(moviedf[doc_column])
    
    
    cosine_sim = cosine_similarity(tfidf_matrix,tfidf_matrix)
    index = moviedf[moviedf["movieId"]==movieId].index[0]
    
    similarity = pd.DataFrame(cosine_sim[index], columns=["similarity"])\
    .sort_values("similarity", ascending=False).head(10).index

    recommend = moviedf['title'].iloc[similarity]
    
    return recommend
```

```python
movies = pd.read_csv("/kaggle/input/the-movies-dataset/movies_metadata.csv",
                    usecols=["title","vote_average","release_date","overview","id"])
ratings = pd.read_csv("/kaggle/input/the-movies-dataset/ratings_small.csv")
```
