<h3 align="left"><font color='#4E5672'>ğŸ› ï¸ Apriori AlgoritmasÄ±</font></h3>
    
* Apriori algoritmasÄ±, genellikle market sepeti analizi iÃ§in kullanÄ±lÄ±r.

* Mlxtend kÃ¼tÃ¼phanesinde bulunan Apriori algoritmasÄ±, "destek" (support) metriÄŸi kullanarak belirlenen destek deÄŸerinin Ã¼zerindeki  Ã¶rÃ¼ntÃ¼leri ve iliÅŸkileri belirler.


```python
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
frequent_itemsets = apriori(basket,min_support=0.1,use_colnames=True,verbose=1)
frequent_itemsets.sort_values("support", ascending=False)
```

```python
rules = association_rules(frequent_itemsets,metric="support",min_threshold=0.01)
rules.sort_values(by="lift")
```

<h3 align="left"><font color='#4E5672'>ğŸ› ï¸ Association Rules AlgoritmasÄ± </font></h3>
    
* Birliktelik kurallarÄ± Ã§Ä±karmak iÃ§in kullanÄ±lan bir algoritmadÄ±r. 
    
* Apriori gibi bir birliktelik analizi algoritmasÄ± tarafÄ±ndan elde edilen kÃ¼melerin **itemset** deÄŸerlerindeki Ã¶ÄŸeler arasÄ±ndaki iliÅŸkilerden kurallar Ã¼retmek amacÄ±yla kullanÄ±lÄ±r.
    

1. **antecedents:** KurallarÄ±n sol tarafÄ±ndaki Ã¶ÄŸeler. (sebepler)
2. **consequents:** KurallarÄ±n saÄŸ tarafÄ±ndaki Ã¶ÄŸeler. (sonuÃ§lar)
3. **antecedent support:** Antecedent'in destek deÄŸeri.
4. **consequent support:** Consequent'Ä±n destek deÄŸeri.
5. **support:** Antecedent ve consequent'Ä±n bir arada gÃ¶rÃ¼lme olasÄ±lÄ±ÄŸÄ±nÄ± belirten destek deÄŸeri.
6. **confidence:**  "EÄŸer A varsa, B'nin de olma olasÄ±lÄ±ÄŸÄ±"nÄ± Ã¶lÃ§en  metriktir (support / antecedent support)
7. **lift:** Birliktelik kurallarÄ±nÄ±n gÃ¼cÃ¼nÃ¼ ifade eder, nekadar yÃ¼ksekse birliktelik kuralÄ± okadar anlamlÄ± (confidence / consequent support)
8. **leverage:** KuralÄ±n gÃ¼venilirlik dÃ¼zeyini Ã¶lÃ§en deÄŸerdir.
9. **conviction:** Birliktelik kurallarÄ±ndaki Ã¶geler arasÄ±ndaki baÄŸÄ±mlÄ±lÄ±ÄŸÄ± belirtir, nekadar yÃ¼ksekse okadar baÄŸÄ±mlÄ±lar.
10. **zhangs_metric:** Birliktelik kurallarÄ±nÄ±n kalitesini Ã¶lÃ§mek iÃ§in kullanÄ±lÄ±r, nekadar yÃ¼ksek okadar kaliteli.


```python
random_product = rules.sample(1,random_state=45)["antecedents"].explode().iloc[0]
random_product
```

```python
def arl_recommender(rules_df, id, rec=1):
    sorted_rules = rules_df.sort_values("lift", ascending=False)
    recommendation_list = []
    for i, k in enumerate(sorted_rules["antecedents"]):
        for j in list(k):
            if j == id :
                for k in list(sorted_rules.iloc[i]["consequents"]):
                    if k not in recommendation_list:
                        recommendation_list.append(k)

    return recommendation_list[0:rec]
```

```python
arl_recommender(rules,random_product,5)
```

```python
products = pd.read_csv("/kaggle/input/instacart-market-basket-analysis/products.csv")
products.head()
```

```python
def names_of_products(rules_df, bought,recommend = 5):
    
    rec = arl_recommender(rules_df,bought,recommend)

    name_of_rec={}
    bought_name = products[products["product_id"]==bought]["product_name"].iloc[0]
    for i in rec:
        name_of_rec[i] = products[products["product_id"]==i]["product_name"].iloc[0]
    recommend_df = pd.DataFrame(name_of_rec.items(), columns=["product_id","product_name"])
    print(f"Bought: {bought_name}\n")
    return recommend_df
```

```python
names_of_products(rules,random_product,5)
```
