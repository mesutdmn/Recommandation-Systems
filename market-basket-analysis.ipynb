<h3 align="left"><font color='#4E5672'> ❔ Nerede Kullanılır </font></h3>

1. **Tahmin ve Strateji Geliştirme**: Müşterilerin alışveriş alışkanlıkları üzerinden gelecekteki alışverişlerini tahmin etmemizi sağlar, bu sayede marketi günlük/saatlik gibi satışlarımızı arttırcak şekilde düzenleyebiliriz. 
    
    **Örnek**: Haftanın farklı günlerinde, mesela cumartesi günü daha fazla eğlence tüketimi ürünlerini ön plana çıkarabiliriz.

2. **Ürün Yerleşimi ve Raf Düzeni**: Hangi ürünlerin birlikte satın alındığını anlayabilirsek, mağaza içindeki ürün yerleşimini ve raf düzenini kârımızı arttırcak şekilde optimize edebiliriz. 
    
    **Örnek**: kola alanların cips alması, salata malzemesi alanların limon alması, organik sebze alanın, yine organik başka bir sebze alması.

3. **Kampanya ve İndirim Stratejileri**: Hangi ürünlerin birlikte satın alındığını bildiğimizde, birine kampanya yapıp, diğerinin fiyatını ufak bir şekilde arttırarak, kârımızı azaltmadan, birini alanın diğerini alcağını bildiğimizden satışlarımızı birlikte yaparak arttırabiliriz. 
    
    **Örnek**: Kola alımında patates cipsi alımıda artıyor ise, Kola alanlara patates cipsi %50 indirimli gibi indirim stratejisi uygulayabiliriz.    

4. **Stok Yönetimi**: Mevsimsel olarak satışı artan ürünlerin stoğunun düzenlenmesi yada birlikte satılan ürünlerin stoklarınında eşit veya yakın tutulmasını sağlamak için kullanılabilir.
    
    **Örnek**: Yağmurlu dönemlerde şemsiye stoğunun daha fazla tutulması veya Şampuan ve Saç kremi stoğunun yakın tutulması.

<h3 align="left"><font color='#6B8BA0'>👀 Features: </font></h3>
    
Orders datasındaki değişkenlerin anlamları şu şekildedir:

- **order_id**: Her siparişin benzersiz bir kimliğidir.
- **user_id**: Kullanıcının benzersiz bir kimliğidir.
- **eval_set**: Bu değişken, veri kümesinin tipini belirtir. "prior", "train", ve "test" olmak üzere üç farklı değeri alabilir.

    **prior**: Bu değer, veri kümesinin eğitim öncesi (prior) bölümünü temsil eder. Yani, kullanıcıların geçmiş siparişleri bu kategoriye aittir.

    **train**: Bu değer, eğitim veri kümesini temsil eder. Modelin eğitildiği ve öğrenme sürecinin gerçekleştiği bölümdür. Model, bu bölümdeki verileri kullanarak öğrenir.

    **test**: Bu değer, test veri kümesini temsil eder. Modelin eğitildikten sonra performansının değerlendirildiği ve tahminlerin yapıldığı bölümdür. Bu kısımdaki veriler, genellikle modelin daha önce görmediği, tahmin yapması gereken yeni verileri içerir.
- **order_number**: Kullanıcının her siparişi içindeki sırasını belirtir.
- **order_dow**: Siparişin haftanın hangi gününde yapıldığını belirtir (0'dan 6'ya kadar olan bir sayı, Pazar'dan Cumartesi'ye).
- **order_hour_of_day**: Siparişin gün içinde hangi saatte yapıldığını belirtir (0'dan 23'e kadar olan bir sayı).
- **days_since_prior_order**: Önceki siparişten bu siparişe kadar geçen gün sayısını belirtir. Eğer bu bir kullanıcının ilk siparişi ise NaN (Not a Number) olabilir.

<h3 align="left"><font color='#4E5672'>🛠️ Apriori Algoritması</font></h3>
    
* Apriori algoritması, genellikle market sepeti analizi için kullanılır.

* Mlxtend kütüphanesinde bulunan Apriori algoritması, "destek" (support) metriği kullanarak belirlenen destek değerinin üzerindeki  örüntüleri ve ilişkileri belirler.


```python
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
frequent_itemsets = apriori(basket,min_support=0.1,use_colnames=True,verbose=1)
frequent_itemsets.sort_values("support", ascending=False)
```

```python
rules = association_rules(frequent_itemsets,metric="support",min_threshold=0.01)
rules.sort_values(by="lift")
```

<h3 align="left"><font color='#4E5672'>🛠️ Association Rules Algoritması </font></h3>
    
* Birliktelik kuralları çıkarmak için kullanılan bir algoritmadır. 
    
* Apriori gibi bir birliktelik analizi algoritması tarafından elde edilen kümelerin **itemset** değerlerindeki öğeler arasındaki ilişkilerden kurallar üretmek amacıyla kullanılır.
    

1. **antecedents:** Kuralların sol tarafındaki öğeler. (sebepler)
2. **consequents:** Kuralların sağ tarafındaki öğeler. (sonuçlar)
3. **antecedent support:** Antecedent'in destek değeri.
4. **consequent support:** Consequent'ın destek değeri.
5. **support:** Antecedent ve consequent'ın bir arada görülme olasılığını belirten destek değeri.
6. **confidence:**  "Eğer A varsa, B'nin de olma olasılığı"nı ölçen  metriktir (support / antecedent support)
7. **lift:** Birliktelik kurallarının gücünü ifade eder, nekadar yüksekse birliktelik kuralı okadar anlamlı (confidence / consequent support)
8. **leverage:** Kuralın güvenilirlik düzeyini ölçen değerdir.
9. **conviction:** Birliktelik kurallarındaki ögeler arasındaki bağımlılığı belirtir, nekadar yüksekse okadar bağımlılar.
10. **zhangs_metric:** Birliktelik kurallarının kalitesini ölçmek için kullanılır, nekadar yüksek okadar kaliteli.


```python
random_product = rules.sample(1,random_state=45)["antecedents"].explode().iloc[0]
random_product
```

```python
def arl_recommender(rules_df, id, rec=1):
    sorted_rules = rules_df.sort_values("lift", ascending=False)
    recommendation_list = []
    for i, k in enumerate(sorted_rules["antecedents"]):
        for j in list(k):
            if j == id :
                for k in list(sorted_rules.iloc[i]["consequents"]):
                    if k not in recommendation_list:
                        recommendation_list.append(k)

    return recommendation_list[0:rec]
```

```python
arl_recommender(rules,random_product,5)
```

```python
products = pd.read_csv("/kaggle/input/instacart-market-basket-analysis/products.csv")
products.head()
```

```python
def names_of_products(rules_df, bought,recommend = 5):
    
    rec = arl_recommender(rules_df,bought,recommend)

    name_of_rec={}
    bought_name = products[products["product_id"]==bought]["product_name"].iloc[0]
    for i in rec:
        name_of_rec[i] = products[products["product_id"]==i]["product_name"].iloc[0]
    recommend_df = pd.DataFrame(name_of_rec.items(), columns=["product_id","product_name"])
    print(f"Bought: {bought_name}\n")
    return recommend_df
```

```python
names_of_products(rules,random_product,5)
```

<h3 align="left"><font color='#4E5672'>🛠️ H-Mine Algoritması</font></h3>
    
* H-Mine (Hash-Based Mining) algoritması, Apriori gibi ilişkisel kural madenciliğinde kullanılan büyük veri kümeleri üzerinde hızlı bir şekilde çalışan bir sıklık tabanlı madencilik algoritmasıdır.
    
* Veri kümesini tarar ve sıklık tabanlı filtreleme uygular.
    
* Hash tabanlı bir yapı kullanır, bu nedenle bellek kullanımı daha etkili olabilir.


```python
from mlxtend.frequent_patterns import hmine
```

```python
frequent_itemsets = hmine(basket,min_support=0.05,use_colnames=True)
frequent_itemsets.sort_values("support", ascending=False)
```
